```python
# Synthetic dataset
from sklearn.datasets import make_classification
# Data processing
import pandas as pd
import numpy as np
from scipy.io import loadmat
from collections import Counter
# Visualization
import matplotlib.pyplot as plt
# Model and performance
from sklearn.svm import OneClassSVM
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
# Load the lympho.mat dataset
data = loadmat('C:/Users/Gokul/Documents/Python Scripts/lympho.mat')
```

```python
df = pd.DataFrame(data['X'], columns=['feature{}'.format(i) for i in range(1,19)])
```

```python
data['X'].shape
```

```python
df['target'] = data['y']
```

```python
df
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data['X'], data['y'], test_size=0.3, random_state=42)
```

```python
# Print the shapes of the resulting arrays
print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)
print('X_test shape:', X_test.shape)
print('y_test shape:', y_test.shape)
```

```python
# Train the one class support vector machine (SVM) model
one_class_svm = OneClassSVM(nu=0.01, kernel = 'rbf', gamma = 'auto').fit(X_train)
```

```python
# Predict the anomalies
prediction = one_class_svm.predict(X_test)
# Change the anomalies' values to make it consistent with the true values
prediction = [1 if i==-1 else 0 for i in prediction]
```

```python
# Check the model performance
print(classification_report(y_test, prediction))
```

```python
print("The accuracy score predicted is {}".format(accuracy_score(y_test, prediction)))
```
